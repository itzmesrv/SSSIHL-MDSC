Stacked Bar Plot

It is a type of bar plot that displays multiple categories on top of one another within each bar.<br>
Height and length represents the value of specific category.
Stacked Bar Plot is used to emphasize both the inidvidual values and the overall composition.

```python
# Resource: Shared PPT
species = (
    "Adelie\n $\\mu=$3700.66g",
    "Chinstrap\n $\\mu=$3733.09g",
    "Gentoo\n $\\mu=5076.02g$",
)
weight_counts = {
    "Below": np.array([70, 31, 58]),
    "Above": np.array([82, 37, 66]),
}
width = 0.5

fig, ax = plt.subplots()
bottom = np.zeros(3)

for boolean, weight_count in weight_counts.items():
    p = ax.bar(species, weight_count, width, label=boolean, bottom=bottom)
    bottom += weight_count

ax.set_title("Number of penguins with above average body mass")
ax.legend(loc="upper right")

plt.show()
```

Box Plot

Box plot summarize data values and its created from first quartile to third quartile<br>
The vertical line which goes through the box represents the median.<br>
Box plot represents data on the x-axis and frequency distribution on the y-axis.<br>
Box plot displays properties like minimum, first quartile, median, third quartile and maximum<br>
The whiskers represents the minimum and maximum: down/left whisker length is more - left skewed & up/right whisker length is more - right skewed<br>
Some of the Params of boxplot are:<br>
The notch - True attribute which creates the notch format to the box plot<br>
patch_artist - True fills the boxplot with colors<br>
The vert = 0 attribute creates horizontal box plot

```python
# definig a function which displays all the below plots for all attributes in the dataframe
def diagnostic_plots(df, variable,target):
    plt.figure(figsize=(20, 7))

    # lineplot
    plt.subplot(1, 5, 1)
    plt.plot(df[variable],color='g')
    plt.title('Lineplot')
    
    # histogram
    plt.subplot(1, 5, 2)
    sns.histplot(df[variable], kde = True,color='g')
    plt.title('Histogram')


    # scatterplot
    plt.subplot(1, 5, 3)
    plt.scatter(df[variable],df[target],color = 'b')
    plt.title('Scatterplot')
    
    # boxplot
    plt.subplot(1, 5, 4)
    sns.boxplot(y=df[variable],color = 'r')
    plt.title('Boxplot')
    
    # barplot
    plt.subplot(1, 5, 5)
    sns.barplot(x = target, y = variable, data = df,color = 'g')   
    plt.title('Barplot')
    
    plt.show()

for col in df:
    diagnostic_plots(df,col,'Outcome')
```
