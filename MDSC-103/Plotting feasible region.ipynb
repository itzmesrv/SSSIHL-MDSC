```python
# importing required libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from scipy.optimize import LinearConstraint
```

Max z = 5x1 + 4x2<br>
Subject to<br>
6x1+4x2<=24<br>
x1+2x2<=6<br>
x2-x1<=1<br>
x2<=2<br>
x1,x2>=0

```python
linear_constraint = LinearConstraint([[6, 4], [1, 2], [-1, 1], [0, 1]], [-np.inf, 1, -np.inf, 1], [24, 6, 1, 2])
```

```python
def fun(x):
    c = np.array([-5, -4]) 
    return np.dot(c, x) 
```

```python
x0 = np.array([0, 0])
result = minimize(fun, x0, constraints=linear_constraint)
```

```python
print("Solution:", result.x)
print("Optimal Value:", result.fun)
```
