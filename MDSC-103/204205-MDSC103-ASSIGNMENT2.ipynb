```python
# importing required libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# defining the given function
f = lambda x: x**2
x = np.linspace(-5,5,100)

# plotting the function
plt.plot(x,f(x))
plt.title('y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# assigning values
x0 = 4
eta = 0.1
delta = 1e-5

# gradf function
def gradf(x):
    return 2*x

x1 = x0 - eta*gradf(x0)

a = []

# performing 100 iterations
for i in range(0,100):
    x1 = x0 - eta*gradf(x0)
    x0 = x1
    if gradf(x0) < delta:
        break
    a.append(x1)

print(x1)

# graph plot with x,f(x) values with the function plot
plt.scatter(a,f(np.array(a)))
plt.plot(x,f(x))
plt.title('Plot of all x,f(x) values')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
    
```

```python
# defining a function to gradient descent for given values in parameters
def gd_function(f,gradf,x0,delta,eta):
    x_val = [x0]
    fun_val = [f(x0)]
    x1 = x0 - eta * gradf(x0)
    count = 0
    
    # condition for while loop to terminate
    while abs(x1 - x0) > delta:
        fun_val.append(f(x1))
        x0 = x1
        x1 = x0 - eta * gradf(x0)
        x_val.append(x1)
        count = count + 1
    
    # plotting the graph with x and f(x) values along with the function plot
    plt.scatter(x_val,fun_val)
    plt.plot(x,f(x))
    plt.title('Plotting values using Gradient Descent Function')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.show()

gd_function(f,gradf,4,1e-5,0.1)
```
