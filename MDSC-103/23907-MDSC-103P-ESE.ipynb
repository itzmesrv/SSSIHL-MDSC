```python
import matplotlib.pyplot as plt
import numpy as np

f = lambda x: 4*x[0] + 6*x[1] - 2*x[0]**2 - 2*x[0]*x[1] - 2*x[1]**2 
x1 = np.linspace(-1,5,100)
x2 = np.linspace(-1.5,1.5,100)
arr = np.array(np.meshgrid(x1,x2))
F = f(arr)
fig = plt.figure(figsize=(15,10)) # figure size
ax = plt.subplot(1,2,1,projection='3d') # 3D plot axis
ax.scatter3D(1,1,f([1,1]),c="black",marker="*",s=100) # using Scatter 3D
surf = ax.plot_surface(arr[0],arr[1],F, cmap= 'gist_rainbow') # surface plot
fig.colorbar(surf,shrink=0.4, aspect=10) # to display the colorbar
ax.set_xlabel('x1', labelpad=10) # x label - x1
ax.set_ylabel('x2', labelpad=30) # y label - x2
ax.set_zlabel('f(X)', labelpad=10) # z label - f(x)
plt.show() # displaying
```
