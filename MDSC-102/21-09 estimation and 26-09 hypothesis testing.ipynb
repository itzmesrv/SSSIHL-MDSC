```python
# reading the dataset and printing it
df = pd.read_csv("E:/Siva/SSSIHL/MSc Data Science/1st Sem/102 Lab/wine/wine.data")

# giving names to all columns as attributes in the dataframe
attributes = ['Class','Alcohol','Malic Acid','Ash','Alcanlinity of ash','Magnesium','Total Phenols','Flavanoids','Nonflavanoid phenols','Proanthocyanins','Color Intensity','Hue','OD280/OD315 of diluted wines','Proline']

df = pd.read_csv("E:/Siva/SSSIHL/MSc Data Science/1st Sem/102 Lab/wine/wine.data",names = attributes)

# giving names to all columns as attributes in the dataframe
df.columns = attributes

# printing the dataframe
df
```

```python
int_columns = df.select_dtypes(include=['int']).columns
df = df.drop(int_columns, axis=1)
```

```python
# Using boxcox and finding the lambda value for all the attributes
transformed_df = pd.DataFrame()
lam_values = {}

for feature in df.columns:
    x = df[feature]
    arr, lam = scipy.stats.boxcox(x)
    transformed_df[feature] = arr
    lam_values[feature] = lam

for feature, lam in lam_values.items():
    print("{}: {}".format(feature, lam))
```

```python
# Applying boxcox transformation i.e. (x^lam - 1) / lam and finding the transformed skewness
transformed_skewval = {}
new_df = pd.DataFrame()

for column in df.columns:
    lam = lam_values[column]
    transformed_data = ( (df[column] ** lam) - 1 ) / lam
    new_df[column] = transformed_data
```

```python
# To estimate Âµ

import math
alpha = 0.05
interval_estimation_mean = {}

for column in new_df.columns:
    mean = new_df[column].mean()
    n = len(new_df)
    sample_variance = new_df[column].var()
    standard_deviation = math.sqrt(sample_variance)
    t_value = stats.t.ppf(1 - (1 - alpha) / 2, n - 1)
    rhs = t_value * standard_deviation
    lb = mean - rhs
    ub = mean + rhs
    interval_estimation_mean[column] = (lb, ub)
```

```python
for column, interval in interval_estimation_mean.items():
    print(f'{column} CI: ({interval[0]}, {interval[1]})')
```

```python
# Using Standard Normal Distribution

alpha = 0.05
interval_estimation_z = {}

for column in new_df.columns:
    mean = new_df[column].mean()
    n = len(new_df)
    sample_variance = new_df[column].var()
    standard_deviation = math.sqrt(sample_variance)
    z_value = stats.norm.ppf(1 - (1 - alpha) / 2)
    rhs = z_value * standard_deviation
    lb = mean - rhs
    ub = mean + rhs
    interval_estimation_z[column] = (lb, ub)
```

```python
for column, interval in interval_estimation_z.items():
    print(f'{column} CI: ({interval[0]}, {interval[1]})')
```

```python
# To estimate Variance

alpha = 0.05
interval_estimation_var = {}

for column in new_df.columns:
    mean = new_df[column].mean()
    n = len(new_df)
    sample_variance = new_df[column].var()
    chi_lower = stats.chi2.ppf(alpha / 2, n-1)
    chi_upper = stats.chi2.ppf(1 - alpha / 2, n-1)
    lb = ((n - 1) * sample_variance) / chi_upper
    ub = ((n - 1) * sample_variance) / chi_lower
    interval_estimation_var[column] = (lb, ub)

print('chi_lower',chi_lower)
print('chi_upper',chi_upper)
```

```python
for column, interval in interval_estimation_var.items():
    print(f'{column} CI: ({interval[0]}, {interval[1]})')
```

```python
import math
h0 = 5.2
alpha = 0.05
mean = sample.mean()
var = 4
n = 10

z_cal = (mean - h0 * math.sqrt(n)) / var
z = stats.norm.cdf(np.abs(z_cal))
p = 2 * (1-z)
if p < alpha:
    print('Reject')
else:
    print('Donot reject')
```
