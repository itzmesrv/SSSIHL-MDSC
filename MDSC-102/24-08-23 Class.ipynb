```python
# Transformation of Magnesium into squareroot
sq_root = np.sqrt(df['Magnesium'])
sq_root.skew()
```

```python
# Transformation of Magnesium into log
ln = np.log(df['Magnesium'])
ln.skew()
```

```python
# Transformation of Magnesium into 1/x
by = 1 / df['Magnesium']
by.skew()
```

```python
# Using boxcox and finding the lambda value
x = df['Magnesium']
arr, lam = scipy.stats.boxcox(x)
lam
```

```python
# Applying boxcox transformation i.e. (x^lam - 1) / lam
boxcox = ( (df['Magnesium'] ** lam) - 1 ) / lam
boxcox.skew()
```

```python
# Plotting the histplot
sns.histplot(x=boxcox,kde=True,color='orange')
plt.title('Histplot for Magnesium which is transformed by boxcox')
```

```python
# Subplots for qqplot - probability plot known as ppplot in scipy
fig, axes = plt.subplots(1,2,figsize=(15,5)) # required rows,cols and plotsize
fig.subplots_adjust(hspace=0.3, wspace=0.3)

stats.probplot(x,plot=axes[0])
stats.probplot(boxcox,plot=axes[1])

axes[0].set_title('Probability Plot for Magnesium')
axes[1].set_title('Probability Plot for transformed magnesium - Boxcox')
```
