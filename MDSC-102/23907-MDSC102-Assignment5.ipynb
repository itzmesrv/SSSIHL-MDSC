```python
# importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from matplotlib.pyplot import figure
```

```python
# reading the dataset and printing it
df = pd.read_csv("E:/Siva/SSSIHL/MSc Data Science/1st Sem/102 Lab/wine/wine.data")
df
```

```python
# giving names to all columns as attributes in the dataframe
attributes = ['Class','Alcohol','Malic Acid','Ash','Alcanlinity of ash','Magnesium','Total Phenols','Flavanoids','Nonflavanoid phenols','Proanthocyanins','Color Intensity','Hue','OD280/OD315 of diluted wines','Proline']
df.columns = attributes
df
```

```python
# Count plot for the first attribute - Class
sns.countplot(data=df,x='Class')
plt.ylabel('Count')
plt.title('Count plot for Class attribute')
```

```python
# Scatterplot between Alcohol and Flavanoids basing on their classes
sns.scatterplot(data=df,x='Alcohol',y='Flavanoids',hue='Class',palette=['red','blue','green'])
plt.title('Scatterplot for Alcohol and Flavanoids based on their classes')
```

```python
# Grid plot for Magnesium vs all other attributes basing on their Classes using scatterplot
fig, axes = plt.subplots(4,3,figsize=(20,20)) # required rows,cols and plotsize
fig.subplots_adjust(hspace=0.3, wspace=0.3) # required height and width space between two sub plots

# for loop to compare Magnesium with all the other attributes basing on their Classes
i=j=0
for feature in df.columns:
    if j > 2:
        j = 0
        i += 1
    
    if feature != 'Magnesium' and feature != 'Class':  # if statement which doesn't conisder magnesium or class as a value
        sns.scatterplot(data=df,x='Magnesium',y=feature,hue='Class',ax=axes[i,j],palette=['red','blue','green'])
        axes[i,j].set_title('Magnesium vs {} '. format(feature))
        j +=1
```

Inferences made from the above subplots (Plots row wise)

Plot 1: Magnesium vs Alcohol - Alcohol content is high for Class1 and less for Class3\
Plot 2: Magnesium vs Malic Acid - Magnesium is mostly in the range of 90-120 for Class1\
Plot 3: Magnesium vs Ash: Magnesium content in Class2 is widely spread than the other classes\
Plot 4: Magnesium vs Alcanlinity of ash - Class3 is optimum when compared to other classes\
Plot 5: Magnesium vs Total Phenols - Total Phenols in Class1 are higher compared to other classes\
Plot 6: Magnesium vs Flavanoids - Flavanoids in Class1 are more compared to other classes\
Plot 7: Magnesium vs Nonflavanoid phenols - Magnesium in Class2 is widely spread when compared to other classes\
Plot 8: Magnesium vs Proanthocyanins - Magnesium in Class3 is in the range 80-120 whereas it is 120+ for other classes\
Plot 9: Magnesium vs Color Intensity - Color Intensity is high for Class3 than other classes\
Plot 10: Magnesium vs Hue - Magnesium in Class2 is widely spread than other classes\
Plot 11: Magnesium vs OD280/OD315 of diluted wines - OD280/OD315 is very less in Class3 than other classes\
Plot 12: Magnesium vs Proline - Proline is high in Class1 than other classes

```python
# Grid plot for Magnesium basing on their individual classes using histogram
fig, axes = plt.subplots(3,1,figsize=(40,40)) # required rows,cols and plotsize
fig.subplots_adjust(hspace=0.3, wspace=0.3) # required height and width space between two sub plots

def histplot(df,i):
    sns.histplot(ax=axes[i],kde=True,data=df,x='Magnesium')
           
histplot(df[df['Class']==1],0)
histplot(df[df['Class']==2],1)
histplot(df[df['Class']==3],2)
```

```python
# Grid plot for all attributes except Magnesium basing on their individual classes using histogram
fig, axes = plt.subplots(12,3,figsize=(20,60)) # required rows,cols and plotsize
fig.subplots_adjust(hspace=0.3, wspace=0.3) # required height and width space between two sub plots

def histplot(df,i):
    j = 0
    #for loop which iterates through every feature
    for feature in df.columns:
        if feature != 'Class' and feature != 'Magnesium':
            sns.histplot(ax=axes[i,j],kde=True,data=df,x=feature)
            j = j + 1
        if j > 2:
            j = 0
            i = i + 1
            
histplot(df[df['Class']==1],0)
histplot(df[df['Class']==2],4)
histplot(df[df['Class']==3],8)
```
