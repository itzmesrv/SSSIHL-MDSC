Importing required libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

Binomial Distribution

```python
pdf = lambda n,p,x: math.comb(n,x) * p ** x * (1-p) ** (n-x)

def plot(n,p):
    res = [pdf(n,p,x) for x in range(n+1)]

    plt.bar(range(n+1),res,color='green')
    plt.xlabel('Number of Successes')
    plt.ylabel('Probability')
    plt.title('Binomial Distribution (n={},p={})'.format(n,p))

plot(30,0.2)
```

Poisson Distribution

```python
pdf = lambda lambd,x: math.exp(-lambd) * (lambd ** x) / math.factorial(x)

def plot(lambd,max):
    res = [pdf(lambd,x) for x in range(max+1)]

    plt.bar(range(max+1),res,color='red')
    plt.xlabel('Number of events')
    plt.ylabel('Probability')
    plt.title('Poisson Distribution (λ={})'.format(lambd))

plot(5,20)
```

Exponential Distributions

```python
pdf = lambda lambd, x: lambd * np.exp(-lambd * x)

def plot(lambd,max):
    x = np.linspace(0, max, 100)
    res = pdf(lambd,x)

    plt.plot(x,res)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Exponential Distribution')

plot(0.5,10)
plot(0.9,8)
plot(1.15,15)
plot(1.5,12)
```

Normal Distribution

```python
pdf = lambda mean, std, x: (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std) ** 2)

def plot(mean,std,xmin,xmax):
    xval = np.linspace(xmin,xmax,100)
    res = pdf(mean,std,xval)

    plt.plot(xval,res)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Normal Distribution')

plot(0,1,-5,5)
plot(0.5,2,-10,10)
plot(2,1,-7,7)
```

Gamma Distribution

```python
pdf = lambda alpha, lambd, x: (x ** (alpha - 1) * np.exp(-x / lambd)) / (math.factorial(alpha-1) * lambd ** alpha)

def plot(alpha,lambd,xmin,xmax):
    xval = np.linspace(xmin,xmax,100)
    res = pdf(alpha,lambd,xval)

    plt.plot(xval,res)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Gamma Distribution')

plot(2,1,0,10)
plot(3,2,0,10)
```
